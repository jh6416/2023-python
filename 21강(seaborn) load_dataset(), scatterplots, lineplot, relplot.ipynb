```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

# load_dataset

seaborn 라이브러리에 존재하는 데이터를 가지고 오는 명령어

```python
tips=sns.load_dataset("tips")
```

```python
tips.head(20)
```

# scatterplots

```python
sns.set_theme()
sns.scatterplot(data=tips, x="total_bill", y= "tip")
```

위 그래표가 나타내는 것은
total_bill이 늘어날수록 tip도 대략 늘어나는

양의 상관관계의 그래프이다.

```python
plt.scatter(tips["total_bill"],tips["tip"])
plt.xlabel("total_bill")
plt.ylabel("tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker")
```

```python
sns.scatterplot(data=tips, x="total_bill", y= "tip", hue="day")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="time")
```

```python
sns.scatterplot(data=tips, x="total_bill",y="tip",hue="sex",style="sex")
```

# relplot : 데이터를 다양한 기준으로 따로 그래프화 하는 메소드

### col = "" : 가로로 구분
### row = "" : 세로로 구분

```python
sns.relplot(data=tips, x="total_bill", y="tip",hue="smoker",col="day", row="sex",
           aspect=1,
           height=2)
```

```python
sns.histplot(data=tips, x="tip")
```

```python
tips["tip"].value_counts()
```

```python
sns.histplot(data=tips, y="tip")
```

```python
sns.relplot(data=trips, 
            x="hour", 
            y="total",
            kind="line",
            hue="payment",
            col="pickup_borough",
            aspect=1
           )
```
