```python
died=[10,25,5,30,21]
germinated=[74,88,56,69,59]
seed = ["Spinach","Turnip","Rhubard","Broccoli","Kale"]
```

```python

plt.bar(seed,germinated, color="#00FF00")
plt.title("Germination Rates")
plt.ylabel("Number Of Seeldings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(seed,germinated, color="orchid")
plt.title("Germination Rates")
plt.ylabel("Number Of Seedings")
```

```python
plt.figure(figsize=(6,6))
plt.bar(seed,died, color="b")
plt.bar(seed,germinated,color="#32CD32",align="edge",width=0.4)
```

```python
plt.figure(figsize=(6,6))
plt.bar(seed,died, color="b",bottom=germinated,label="Failed Seedlings")
plt.bar(seed,germinated,color="#32CD32",label="Germinated Seedlings")
plt.title("Greenhouse Seedlings")
plt.legend()
```
