```python
import csv
import matplotlib.pyplot as plt

f = open("subway202306.csv")
data = csv.reader(f)
header1 = next(data)
header2 = next(data)

mx = ["0"] * 23
mx_station=[""]*23

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(23):
        a = row[j*2+4]
        if a > int(mx[j]):
            mx[j] = a
            mx_station[j] = row[3]
            
dic = { y:x for x, y in zip(mx_station,mx) }

plt.rc("font", family="Malgun Gothic")
plt.bar(range(23), mx)
plt.xticks(range(23), mx_station, rotation=90)
plt.xlabel("역이름")
plt.ylabel("인구수")
plt.text(dic[],max(dic),
         'Min Point',
         color='r',
         horizontalalignment='center',
         verticalalignment='top')
plt.show()
```
