문제 2 : 아래 데이터를 입력하고 보이는 그래프를 표출하는 코딩을 하세요.

```python
height = [137,140,142,145,147,150,152,155,157,160]
f_weights = [28.5,30.8,32.6,34.9,36.4,39,40.8,43.1,44.9,47.2]
m_weights = [34.9,38.1,33.5,35.8,46.7,42.8,43.1,45.8,50.8,58.9]

plt.style.use("ggplot")
plt.plot(height,f_weights,"r-*",label="Female")
plt.plot(height,m_weights,"b-d",label="Male")
plt.xlabel("Height(cm)")
plt.ylabel("Weight(kg)")
plt.legend()
plt.show()
```

문제 4 : 아래와 같이 titanic pclass별로 탑승자 연령분포도를 그리세요

```python
titanic["pclass"]=pd.to_numeric(titanic["pclass"], errors="coerce")

first_class = titanic[titanic["pclass"]==1]["age"]
second_class = titanic[titanic["pclass"]==2]["age"]
third_class = titanic[titanic["pclass"]==3]["age"]

plt.figure(figsize=(10,4))
ax = plt.subplot(1,3,1)
plt.hist(first_class,label="1st Class",color="yellow",bins=20)
plt.title("1st Class")
ax1 = plt.subplot(1,3,2,sharey=ax)
plt.hist(second_class,label="2st Class",color="blue",bins=20)
plt.title("2st Class")
ax2 = plt.subplot(1,3,3,sharey=ax)
plt.hist(third_class,label="3st Class",color="red",bins=20)
plt.title("3st Class")
```

문제 5 : 아래와 같이 하루 일정표를 pie 그래프로 나타내시요.

```python
label = [
    "sleeping",
    "playing with friends",
    "eating",
    "exercising",
    "studying"
]
time = [
    5.74,
    3.79,
    3.02,
    2.25,
    7.5
]

plt.pie(time,
        labels=label,
        explode=(0,0.1,0,0,0),
        
        autopct="%.1f%%")
plt.show()
```

문제 6 : kc_house_data.csv 데이터를 활용하여 각 주택의 거실의 크기에 따른 부지 크기를 함께 나타내도록 그래프를 그리세요

문제 7 : nuforc_reports.csv 데이터를 활용하여 월별 UFO 출몰 빈도를 나타내는 그래프를 그리세요.

문제 8 : Salaries.csv 데이터에서 앞에서부터 7명의 기본금, 초과근무수당, 기타수당을 한 그래프에 나타내도록 코딩하세요.

```python
salarie = pd.read_csv("data/salaries.csv", low_memory=False)
salarie["BasePay"]=pd.to_numeric(salarie["BasePay"], errors="coerce")
salarie["OvertimePay"]=pd.to_numeric(salarie["OvertimePay"], errors="coerce")
salarie["OtherPay"]=pd.to_numeric(salarie["OtherPay"], errors="coerce")

df = salarie[["EmployeeName", "BasePay","OvertimePay","OtherPay"]]
df.set_index("EmployeeName")
df.head(7).plot(kind="bar", stacked=True)
```

문제 9 : Salaries.csv 데이터를 활용하여 초과근무수당이 가장 높은 순서대로 TOP 10 에 해당하는 데이터가 그래프로 표출되도록 코딩하세요

```python
df["BasePay"].sort_index(ascending=False).head(7).plot(kind="bar")
```
