```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

위 명령어는 matplotlib이 현재 주피터노트북에서 실행될 수 있도록 한다. 

```python
x= np.linspace(0,5,11)
x
```

```python
y=x**2
y
```

```python
c=x**3
c
```

```python
plt.plot(x,y,"#2ae2e2")
plt.xlabel("X Axis Title Here")
plt.ylabel("Y Axis Title Here")
plt.title("String Title Here")
plt.show
```

plt matplotlib 을 저장했고, matplotlib에 있는 plot이라는 메소드를 통해 어떤 상관관계가 있는 그래프를 그릴 것인지를 지정하고, 해당 그래프의 색깔을 지정한다.

plot x축에 대입할 값, y축에 대입할 값, "라인 컬러"

plot : 이야기

```python
plt.plot(y,x,"g")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("String Title Here")
plt.show
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")
plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

Subplot (1,2,1)의 의미는, 기존 plot을 1을 기준으로 가로에 두개의 그래프를 넣는다는 의미이고, 그 다음 1은 두개의 그래프 위치 중 첫번째에 plt.plot(x, y, 'r--')  라는 그래프를 넣는다는 의미이다

```python
plt.subplot(2,3,1)
plt.plot(x,y,"r--")
plt.subplot(2,3,3)
plt.plot(y,x,'g*-')
```

```python
plt.subplot(2,3,2)
plt.plot(x,y,"r--")
plt.subplot(1,3,1)
plt.plot(y,x,'g*-')
```

plt.subplot(1,3,1) 은 기존 plot의 범위에 3개의 그래프를 넣는다는 의미이고, 그 그래프 위치 3가지 중 첫번째에 plt.plot(x,y,'r--')그래프를 넣는다는 뜻이다.

```python
plt.subplot(1,4,1)
plt.plot(x,y,'y--')
plt.subplot(1,4,3)
plt.plot(y,x,'b*-')
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.3,0.9])

axes.plot(x,y,"b")
axes.set_xlabel("Set X Label")
axes.set_ylabel("Set Y Label")
axes.set_title("Set Title")
```

Figure라는 메소드는 그래프의 크기와 가로 세로 비율을 정할 수 있도록 해준다.

axes : 축

fig.add_axes([왼쪽축에서의 거리, 바닥에서의 거리, 그래프의 폭, 그래프의 높이])

위 add_axes 안에 들어갈 수 있는 숫자의 범위는 0부터 1까지 가능



```python
fig.add_axes
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) # 정사각형
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # 가로가 넓은 직사각형
axes.plot(x,y,"b")
axes.set_xlabel("X Label_axes2")
axes.set_ylabel("Y Label_axes2")
axes.set_title("Axes 2 Title")
axes2.plot(y,x,"r")
axes2.set_xlabel("X Label_axes2")
axes2.set_ylabel("Y Label_axes2")
axes2.set_title("Axes 2 Title")
```

```python
fig = plt.figure()


axes = fig.add_axes([0.1,0.1,0.8,0.7]) # 정사각형
axes.plot(x,y,"#fb6544")
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("Title")
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
for ax in axes:
    ax.plot(x,y,"b")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title("Title")
fig
```
