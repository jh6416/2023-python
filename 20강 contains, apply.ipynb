파이썬은 대체적으로 위에서 아래로 흐른다.

이와같이 if문을 활용하여 첫번째 반복이 실행되면 첫번째에서 멈추고,
만약 실행되지 않으면 그 밑에 if문으로 내려간다.

만약이 없다면 마지막 else문에서 반복이 끝난다.

```python
def age_group(a):
    if a < 2 :
        return 'infant'
    elif 2<=a<=11:
        return 'child'
    elif 12<=a<=17:
        return 'teen'
    elif 18<= a <=49:
        return 'adult'
    else:
        return 'senior'
```

```python
titanic["age"].apply(age_group).value_counts()
```

```python
titanic["age"].apply(age_group)
```

```python
titanic["age_group"] = titanic["age"].apply(age_group)
```

```python
titanic.groupby("age_group")["survived"].mean()
```

```python
titanic.groupby(["age_group","sex"]).survived.mean()
```

```python
titanic[["pclass","survived","age","fare"]].max()-titanic[["pclass","survived","age","fare"]].min()
```

```python
def get_range(s):
    return s.max() - s.min()
titanic[["pclass","survived", "age", "fare"]].apply(get_range)
```

```python
def get_groups(s):
    if s == 0 :
        return "solo"
    elif s < 5 :
        return "average"
    else:
        return "large"
        
    
titanic["group"] = titanic["sibsp"] + titanic["parch"]
titanic["group"].apply(get_groups)
```
