```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
penguins = sns.load_dataset("penguins")
penguins.head(3)
```

```python
sns.set_theme()
plt.figure(dpi=200)
sns.histplot(data=penguins, x="body_mass_g", bins=20)
```

```python
sns.kdeplot(data=penguins, x="body_mass_g")
```

```python
sns.kdeplot(data=penguins, x="body_mass_g", hue="species", multiple="stack")
```

```python
sns.histplot(data=penguins, x="flipper_length_mm")
```

```python
sns.histplot(data=penguins, x="body_mass_g",y="flipper_length_mm" )
```

1. 펭귄의 몸무게가 늘어날수록 날개 길이가 늘어난다.
2. 펭귄의 몸무게 3700에서 4000 사이, 날개길이 190에서 195사이 데이터가 가장 많이 밀집되어있다.

```python
sns.kdeplot(data=penguins, x="body_mass_g", y="flipper_length_mm")
```

등고선 : 제일 안의 원일수록 데이터가 밀집되어 있다. 밖으로 나갈수록 데이터가 적어지며, 그 외는 데이터가 존재하지 않다.

```python
sns.kdeplot(data=penguins, x="bill_length_mm", y="flipper_length_mm", hue = "species")
```

x는 부리의 길이, y는 날개의 길이를 의미한다.
Adelie가 전반적으로 부리와 날개의 길이가 작고,
Gentoo가 전반적으로 날개의 길이가 크고,
Chinstrap가 전반적으로 부리 길이가 크다.
Adelie는 부리 35 ~ 40사이, 날개 185 ~ 193,
Chinstrap은 부리 48 ~ 53, 날개 195 ~ 200,
Gentoo는 부리 45 ~ 51, 날개 202 ~ 215 정도에서 데이터가 밀집되어 있다.
데이터가 가장 정비례하게 늘어나는 건 Gentoo이다. 즉 부리가 길 때, 날개가 길다고 표현하기 가장 적절한 종이다.
Gentoo의 경우, 부리와 날개의 길이가 짧을 수록 상관관계가 높은 것이 밀집되어 있고,
Chinstrap의 경우, 부리와 날개의 길이가 길수록 상관관계가 높은 것이 밀집되어 있다.
Adelie의 경우, 전반적인 펭귄의 부리길이, 날개길이가 서로 대강 연관되어 있다.

```python
sns.histplot(data=penguins, x="bill_length_mm", y="flipper_length_mm", hue = "species")
```

```python
tips = sns.load_dataset("tips")
```

```python
tips
```

```python
sns.rugplot(data=tips, x="tip")
```

```python
sns.rugplot(data=tips, x="tip", height=0.2)
```

X축이 길수록 팁이 높다. 2~4정도의 팁이 가장 많아 빼곡하게 채워져 있다. 뒤로 갈수록 팁 빈도가 줄어 실들이 줄어든다.

```python
sns.rugplot(data=tips, y="tip", height=0.2)
```

```python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill", height=0.1)
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
sns.rugplot(data=tips, x="total_bill", y="tip")
```

scatterplot은 산점도로, rugplot은 선으로 표시한다.
이 그래프는 양의 상관관계이다.
total_bill 10 ~ 20, tip 1.5 ~ 3.8 정도가 데이터가 가장많아
scatter는 점이 많이 뭉쳐있고, rug는 선들이 많이 뭉쳐있다.
위 그래프는 크기가 커질수로 빈도가 적어진다.

```python
sns.displot(data=penguins, kind="hist", x="body_mass_g", hue="sex", col="species")
```

위 그래표는 펭귄의 종을 기준으로 몸무게를 성별로 나눈 그래프이다.
펭귄의 개체수는 Adelie가 가장 많고, 몸무게도 적은 개체가 많다.
Chinstrap은 개체수가 적어 hist 그래프의 단면적이 가장 작다.
이 종은 다양한 몸무게를 분포하고 있다.
제일 무게가 많이 나가는 펭귄은 Gentoo이다. 성별 비율이 가장 적절하다.
대체적으로 암컷 펭귄보다 수컷 펭귄들이 몸무게가 더 많이 나간다.

```python
sns.displot(data=tips, kind="kde", x="tip",col="time", rug=True)
```

위 그래표는 시간을 기준으로한 팁 데이터이다.
사람들은 점심보단 저녁에 더 많이 방문한다.
점심과 저녁은 대부분 2 ~ 4팁을 받는다.
점심보단 저녁에 tip이 많다.
kde가 높을수록 rug가 많이 뭉친다.
