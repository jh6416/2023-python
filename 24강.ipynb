- dropna(), dropna(how = "any") : NaN이 하나라도 포함된 경우, 그 행을 지운다.
- dropna(how = "all") : 모든 컬럼의 value가 없는 행을 지운다.

```python
nba.dropna(how="all", inplace=True)
nba["Salary"].fillna(0, inplace=True)
nba["College"].fillna("None",inplace=True)
nba.head(6)
```

```python
nba.dropna(how="all").dtypes
```

```python
nba.dropna(how="all").info()
```

```python
nba.head(3)
```

```python
nba["Number"]=nba["Number"].astype("int64")
nba["Age"]=nba["Age"].astype("int64")
nba.head(3)
```
