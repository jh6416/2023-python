```python
left = pd.DataFrame({"Key":["K0","K1","K2","K3"],
                    "A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"]})
```

```python
right = pd.DataFrame({"Key":["K0","K1","K2","K4"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,how='inner',on='Key')
```

Left와 right 두가지 데이터 중 내용이 같은 key를 기준으로 두 데이터를 합쳐서 보여줌.

inner라는 조건은, 두개의 데이터 셋에 공통된 부분만 합하여 준다는 뜻.

```python

right
```

```python
pd.merge(left,right,how='inner',on="Key")
```

```python
pd.merge(left,right,how='outer',on="Key")
```

Outer, Join의 특징

합치려는 (Merge 하려는) 데이터셋 key를 모두 가져와서, 기준 컬럼으로 사용하고,

해당 key에 데이터가 없는 부분은 NaN으로 출력되도록 한다.

```python
left1 = pd.DataFrame({"Key1":["K0","K0","K1","K2"],
                    "Key2":["k0","k1","k0","k1"],
                    "A":["A0","A1","A2","A3"],
                    "B":["B0","B1","B2","B3"]})
```

```python
right1 = pd.DataFrame({"Key1":["K0","K1","K1","K2"],
                    "Key2":["k0","k0","k0","k0"],
                    "C":["C0","C1","C2","C3"],
                    "D":["D0","D1","D2","D3"]})
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1,right1,how='inner',on=["Key1","Key2"])
```

inner라는 조건은 default이기 때문에 쓰지 않아도 된다.

```python
pd.merge(left1,right1,how='outer',on=["Key1","Key2"])
```

```python
pd.merge(left1,right1,how='inner',on=["Key1","Key2"])
```
