```python
titanic = pd.read_csv("data/titanic.csv", low_memory = False)

titanic['age']=titanic['age'].replace(["?"],[None]).astype("float")
titanic
```

```python
df = titanic[["sex","pclass","survived","age"]]
gbo = df.groupby(by="sex")
gbo.get_group("male").mean()
```

```python
for i in gbo:
    print(i[0])
    print("----------")
    print(i[1])
```

```python
titanic.groupby("sex")["age"].agg("max","min")
```

```python
titanic.groupby("sex").agg({"age":["max","min"],"pclass":"mean"})
```

```python
def range(r):
    return r.max() - r.min()

titanic.groupby("pclass")["age"].agg(["max","min",range])
```

```python
def count_null(n):
    return n.size - n.count()

sum(titanic.groupby("pclass")['age'].agg(count_null))
```
