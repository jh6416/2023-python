### age_final.csv에 있는 경기도 성남시 수정구 태평1동을 뽑는 명령어

```python
f = open('age_final.csv')
data = csv.reader(f)
header=next(data)
result = []

for row in data:
    if "수정구 태평1동" in row[0]:
        print(row)
```

### age_final.csv에 있는 중구 회현동에서 신생아의 데이터 명령어

```python
f = open('age_final.csv')
data = csv.reader(f)
header=next(data)
result = []

for row in data:
    if "중구 회현동" in row[0]:
        print(row[3])
            
```

### 신도림 지역의 10세 미만 인구수 출력하는 명령어

```python
f = open('age_final.csv')
data = csv.reader(f)
header=next(data)
result = []

for row in data:
    if "신도림" in row[0]:
        for i in row[3:13]:
            result.append(int(i)) 
print(result)
            
```

### 후암동 직역의 인구분포도를 그리는 방법

```python
f = open('age_final.csv')
data = csv.reader(f)
header=next(data)
result = []

for row in data:
    if "후암동" in row[0]:
        for i in row[3:]:
            result.append(int(i)) 
plt.rc('font', family="Malgun Gothic")
plt.plot(result, "blue", label="연령별분포")
plt.title("후암동 인구분포")
plt.legend()
plt.show()
```

### 후암동 인구분포에서 20대 인구 총 수를 구하는 명령어

```python
f = open('age_final.csv')
data = csv.reader(f)
header=next(data)
result = []
name=input('사는 지역: ' )

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i))
plt.rc('font', family="Malgun Gothic")
plt.plot(result, "blue", label="연령별분포")
plt.title(name+" 인구분포")
plt.legend()
plt.show()
```
