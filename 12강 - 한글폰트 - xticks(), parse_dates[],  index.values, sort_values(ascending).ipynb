```python
c=ufos[ufos["state"]=="CA"]["year"].value_counts().sort_index()
n=ufos[ufos["state"]=="NY"]["year"].value_counts().sort_index()

plt.figure()
plt.plot(c,label="CA", color="pink")
plt.plot(n,label="NY", color="brown")
plt.title("UFO Sighting By State")
plt.xlabel("Year")
plt.legend()
```

```python
ufos[ufos["shape"] == "formation"]. year.value_counts().sort_index().plot(label = "formation")
ufos[ufos["shape"] == "oval"]. year.value_counts().sort_index().plot(label = "oval")
ufos[ufos["shape"] == "disk"]. year.value_counts().sort_index().plot(label = "disk")
plt.legend()
plt.title("UFO Sighting By Shape",fontsize=20)
plt.xlabel("Year",fontsize=20)
```

```python
ufos[ufos["shape"] == "formation"]. year.value_counts().sort_index().plot(label = "formation",lw=4,color="skyblue")
ufos[ufos["shape"] == "oval"]. year.value_counts().sort_index().plot(label = "oval",lw=4,color="red")
ufos[ufos["shape"] == "light"]. year.value_counts().sort_index().plot(label = "light",lw=4,color="yellow")
ufos[ufos["shape"] == "fireball"]. year.value_counts().sort_index().plot(label = "fireball",lw=4,color="green")
ufos[ufos["shape"] == "disk"]. year.value_counts().sort_index().plot(label = "disk",lw=4, color="gray")
plt.legend(loc="upper right")
plt.title("UFO Sighting By Shape",fontsize=20)
plt.xlabel("Year",fontsize=20)
```

 Loc 는 location
Upper right 는 그래프의 위쪽, 오른쪽으로 label을 위치하도록 한다.
