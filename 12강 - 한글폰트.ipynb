```python
labels = [
    "공부",
    "먹기",
    "놀고 휴식하기",
    "잠자기",
    "취미생활"
]
time=[
    13,
    3,
    2,
    6,
    1.5
]
colors = ["#355507", "#6d597a", "#d56576","#e56d6f","#eaac8d"]
```
